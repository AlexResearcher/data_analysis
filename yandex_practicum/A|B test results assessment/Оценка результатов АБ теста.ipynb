## Введение

### Краткое описание, цель и задачи

Краткое описание: необходимо провести оценку результатов A/B-теста интернет-магазина. В нашем распоряжении есть датасет с действиями пользователей, техническое задание и несколько вспомогательных датасетов.
Назначение теста: тестирование изменений, связанных с внедрением улучшенной рекомендательной системы.

Цель - провести оценку результатов A/B-теста.

Задачи:
* Оценить корректность проведения теста.
* Проанализировать результаты теста.

### Техническое задание и данные

Техническое задание:

- Название теста: `recommender_system_test`;
- группы: А — контрольная, B — новая платёжная воронка;
- дата запуска: 2020-12-07;
- дата остановки набора новых пользователей: 2020-12-21;
- дата остановки: 2021-01-04;
- аудитория: в тест должно быть отобрано 15% новых пользователей из региона EU;
- назначение теста: тестирование изменений, связанных с внедрением улучшенной рекомендательной системы;
- ожидаемое количество участников теста: 6000.
- ожидаемый эффект: за 14 дней с момента регистрации пользователи покажут улучшение каждой метрики не менее, чем на 10%:
    - конверсии в просмотр карточек товаров — событие `product_page`,
    - просмотры корзины — `product_cart`,
    - покупки — `purchase`.

В нашем распоряжении 4 датасета:

**ab_project_marketing_events.csv** — календарь маркетинговых событий на 2020 год;
Структура файла:
* name — название маркетингового события;
* regions — регионы, в которых будет проводиться рекламная кампания;
* start_dt — дата начала кампании;
* finish_dt — дата завершения кампании.

**final_ab_new_users.csv** — все пользователи, зарегистрировавшиеся в интернет-магазине в период с 7 по 21 декабря 2020 года;
Структура файла:
* user_id — идентификатор пользователя;
* first_date — дата регистрации;
* region — регион пользователя;
* device — устройство, с которого происходила регистрация.

**final_ab_events.csv** — все события новых пользователей в период с 7 декабря 2020 по 4 января 2021 года;
Структура файла:
* user_id — идентификатор пользователя;
* event_dt — дата и время события;
* event_name — тип события;
* details — дополнительные данные о событии. Например, для покупок, purchase, в этом поле хранится стоимость покупки в долларах.

**final_ab_participants.csv** — таблица участников тестов.
* Структура файла:
* user_id — идентификатор пользователя;
* ab_test — название теста;
* group — группа пользователя.

Но для уверенности проведем небольшой тест. Смоделируем 1500 случайных выборок размером 200 из наших данных для группы А, а затем и для группы Б. Посчитаем средние и построим их распрделение и qq plot. Если распределение похоже на нормальное, то все значения будут на диагонали qq plot или очень близко к ней. Если это так, то т-тест можно применять.
Построим qq plot и гистограмму для группы А.

```python
t = pd.Series(
    [users_lifetime[users_lifetime['group']=="A"].groupby('user_id')['event_name'].count().sample(
        200,replace=True).mean() for _ in range(1500)])
t.hist()

plt.ylabel('частота попаданий в интервал значений')

plt.xlabel('значения средних')

plt.title("Гистограмма для частоты средних событий на пользователя для группы А")

plt.show()
```

```python
stats.probplot(t, dist="norm", plot=pylab)
pylab.show()
```

Гистограмма похожа на нормальное распределение, но qq plot показывает расхождения. Поэтому для верности, кажется, лучше использовать критерий Манна-Уитни.\
Проделаем то же самое для Б группы.

```python
t = pd.Series(
    [users_lifetime[users_lifetime['group']=="B"].groupby('user_id')['event_name'].count().sample(
        200,replace=True).mean() for _ in range(1500)])
t.hist()

plt.ylabel('частота попаданий в интервал значений')

plt.xlabel('значения средних')

plt.title("Гистограмма для частоты средних событий на пользователя для группы Б")

plt.show()
```

```python
stats.probplot(t, dist="norm", plot=pylab)
pylab.show()
```

Здесь qq plot показывает наличие еще более сильных выбросов, поэтому остановимся на тесте Манна-Уитни.

Нулевая гипотеза: среднее количество событий на клиента из группы А не отличается значимо от среднего для группы Б.\
Альтернативная гипотеза: есть значимая разница в среднем числе событий на пользователя. 

```python
values_1 = users_lifetime[users_lifetime['group']=="A"].groupby('user_id')['event_name'].count()
values_2 = users_lifetime[users_lifetime['group']=="B"].groupby('user_id')['event_name'].count()

alpha = 0.05
        
print("p-value {}".format(stats.mannwhitneyu(values_1, values_2)[1]))

p_value = stats.mannwhitneyu(values_1, values_2, alternative='two-sided')[1]
    
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: есть статистически значимые различия в средних значениях.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: статистически значимых различий в средних значениях нет.')
```

Среднее значение событий на пользователя для группы А выше, чем для группы Б и эта разница статистически значима.

И тут все также. 4 разных вида покупок. Для верности можно проверить значимость различий средних тестом Манна-Уитни. Тест Стьюдента тут не подойдет, так как у нас слишком большой разброс значений счетов. 

Нулевая гипотеза: средняя сумма покупки клиентов из группы А не отличается значимо от средней стоимости покупки клиентов из группы Б.\
Альтернативная гипотеза: есть значимая разница в средних суммах покупки. 

```python
values_1 = users_lifetime.loc[(users_lifetime['group']=='A')&(users_lifetime['event_name']=='purchase')]['details']
values_2 = users_lifetime.loc[(users_lifetime['group']=='B')&(users_lifetime['event_name']=='purchase')]['details']

alpha = 0.05
        
print("p-value {}".format(stats.mannwhitneyu(values_1, values_2)[1]))

p_value = stats.mannwhitneyu(values_1, values_2, alternative='two-sided')[1]
    
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: есть статистически значимые различия в средних значениях.')
else:
    print('Не получилось отвергнуть нулевую гипотезу: статистически значимых различий в средних значениях нет.')
```

### Строим воронку

Воронку считаем по уникальным пользователям. Первый шаг - логин. Посчитаем таблицу с воронкой для группы А.

```python
funnel_A = users_lifetime.loc[users_lifetime['group']=='A'].groupby(
               'event_name')['user_id'].nunique().sort_values(ascending = False).reset_index()
#funnel_A
```

```python
funnel=[]
for idx in range(len(funnel_A['user_id'])):
        funnel.append(round(funnel_A.loc[idx,'user_id'] / funnel_A.loc[0,'user_id'] * 100,2))    
        
#создаем столбец с воронкой
funnel_A['funnel'] = funnel

funnel_A = funnel_A.reindex([0,1,3,2])
funnel_A
```

### Проверка гипотез

Проверим статистическую разницу долей z-критерием. Для каждой группы у нас будет 3 сравнения с логином. Поэтому применим поправку равную 3. Это не много, поэтому можно использовать Бонферрони.

**Начнем с шага с просмотром страницы о товарах.**

Нулевая гипотеза: доли пользователей перешедшие на шаг просмотра товаров после логина одинаковые в группе А и Б.\
Альтернативная гипотза: доли разные.

```python
def proportion_test(event, idx):
    b_count = funnel_B[funnel_B['event_name']==event]['user_id'][idx]
    a_count = funnel_A[funnel_A['event_name']==event]['user_id'][idx]
    all_b = funnel_B[funnel_B['event_name']=='login']['user_id'].sum()
    all_a = funnel_A[funnel_A['event_name']=='login']['user_id'].sum()

    alpha = 0.05/3
    count = [b_count, a_count]
    nobs = [all_b, all_a]
    stat, p_value = proportions_ztest(count, nobs)

    print('p-значение: ', p_value)

    if p_value < alpha: 
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница.')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными.')
```

```python
proportion_test('product_page', 1)
```

Первый шаг воронки имеет самое большое различие в долях и тест нам говорит, что оно значимо статистически. Вероятность получить случайно такое или еще более значимое различие сильно меньше 0.05/3.

Первый шаг воронки имеет самое большое различие в долях и тест нам говорит, что оно не значимо статистически, значит, могло быть получено слуайно, когда различий на самом деле нет.
